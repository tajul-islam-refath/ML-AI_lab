```python
def bar_chart(feature, df):
    no = df[df['target']==1][feature].value_counts()
    yes = df[df['target']==0][feature].value_counts()
    df = pd.DataFrame([no,yes])
    df.index = ['Heart problem','No heart problem']
    df.plot(kind='bar',stacked=True, figsize=(10,5))
```

```python
x = df.drop(["target"] , axis=1)
x
```

```python
y = df["target"]
y
```

```python
# fetature selection
from sklearn.feature_selection import SelectKBest , f_classif
```

```python
FIT_FEATURES = SelectKBest(score_func=f_classif)
```

```python
FIT_FEATURES.fit(x , y)
```

```python
score_col = pd.DataFrame(FIT_FEATURES.scores_, columns=["score value"])
score_col
```

```python
col_name = pd.DataFrame(x.columns, columns=["Name"])
col_name
```

```python
top_features = pd.concat([col_name,score_col] , axis=1)
top_features
```

```python
top_features.nlargest(8,"score value")
```

```python
top_x_features = x.drop(["age","trestbps","chol","fbs","restecg"], axis=1)
top_x_features
```

```python
# for separate train and test data
from sklearn.model_selection import train_test_split
# To calculate the accuracy score of the model
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
# svm model
from sklearn.svm import SVC
```

```python
xTrain , xTest , yTrain , yTest = train_test_split(top_x_features , y , test_size=.2, random_state=1)
```

```python
#linear model
model = SVC(C= .1, kernel='linear', gamma= 1)
```

```python
model.fit(xTrain , yTrain)
```

```python
model.get_params()
```

```python
predict = model.predict(xTest)
```

```python
print(f"Train score : {model.score(xTrain , yTrain)}")
print(f"Test score : {model.score(xTest , yTest)}")
```

```python
print(classification_report(yTest, predict))
```

```python
from sklearn.model_selection import GridSearchCV

# defining parameter range
param_grid = {'C': [0.1, 1, 10, 100, 1000], 
              'gamma': [1, 0.1, 0.01, 0.001, 0.0001],
              'kernel': ['rbf', 'linear']} 
```

```python
grid = GridSearchCV(SVC(), param_grid, refit = True, verbose = 3)
```

```python
grid.fit(xTrain, yTrain)
```

```python
# print best parameter after tuning
print(grid.best_params_)
  
# print how our model looks after hyper-parameter tuning
print(grid.best_estimator_)
```

```python
print("Train accuracy : ", grid.best_score_*100)
```

```python
print("Test Accuracy after tuning : ", grid.score(xTest, yTest))
```

```python
grid_predictions = grid.predict(xTest)
  
# print classification report
print(classification_report(yTest, grid_predictions))
```

```python
cm=confusion_matrix(yTest,grid_predictions,labels=grid.classes_)
from sklearn.metrics import ConfusionMatrixDisplay
disp=ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=grid.classes_)
disp.plot()
plt.show()
print("Confusion Matrix:\n",confusion_matrix(yTest,grid_predictions))
```

```python
from sklearn.model_selection import learning_curve
train_sizes, train_scores, test_scores = learning_curve(model, xTrain, yTrain, cv=10, scoring='accuracy', n_jobs=-1, train_sizes=np.linspace(0.01, 1.0, 40))
```

```python
    train_mean = np.mean(train_scores, axis=1)
    train_std = np.std(train_scores, axis=1)

    test_mean = np.mean(test_scores, axis=1)
    test_std = np.std(test_scores, axis=1)

    plt.subplots(1, figsize=(10,5))
    plt.plot(train_sizes, train_mean, '--', color="blue",  label="Training score")
    plt.plot(train_sizes, test_mean, color="orange", label="test score")

   
    plt.fill_between(train_sizes, train_mean - train_std, train_mean + train_std, color="#DDDDDD")
    plt.fill_between(train_sizes, test_mean - test_std, test_mean + test_std, color="#DDDDDD")

   
    plt.title("Learning Curve")
    plt.xlabel("Training Set Size"), plt.ylabel("Accuracy Score"), plt.legend(loc="best")
    plt.tight_layout(); 
    plt.show()
```

```python
error_rate_train = 1 - train_mean
error_rate_test = 1 - test_mean
plt.subplots(1, figsize=(10,5))
plt.plot(train_sizes, error_rate_train, '--', color="blue",  label="Training error")
plt.plot(train_sizes, error_rate_test, color="orange", label="Test error")
plt.legend(loc="upper right")
plt.title("Error Rate")
plt.xlabel("Training Set"), plt.ylabel("Error")
```
